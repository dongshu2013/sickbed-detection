```python
%matplotlib inline  

from matplotlib import pyplot as plt
import logging
import numpy as np
import cv2
import time
import os
from fnmatch import fnmatch
```

```python
def m4v_files_iterator(input_dir):
    for root, dirs, files in os.walk(input_dir):
        for f in files:
            if fnmatch(f, '*.m4v'):
                yield os.path.join(root, f)
```

```python
for video_file in m4v_files_iterator('./video/depth/'):
    cap = cv2.VideoCapture(video_file)
    ret = False
    if cap.isOpened():
        ret, frame = cap.read()

    i = 0
    while(ret):
        i = i + 1
        ret, frame = cap.read()
        if i % 200 == 0:
            cv2.imwrite('./images/depth' + str(int(time.time() * 1000)) + '.jpg', frame)

    cap.release()
    cv2.destroyAllWindows()
```
